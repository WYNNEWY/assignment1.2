```python
from sklearn.preprocessing import LabelEncoder
```

```python
test_data = pd.read_csv('adult.test', names=column_names,sep=' *, *', skiprows=1,na_values='?',engine='python')
```

```python
test_data['income'].replace({'<=50K.':0,'>50K.':1},inplace=True)
```

```python
test_data
```

```python
dft=test_data.fillna({'workclass':'private','occupation':'Prof-specialty','nativecountry':'United-States'})
```

```python
dft.drop(['fnlwgt', 'education'], axis=1, inplace=True)
```

```python
XT = dft.drop(['income'],axis=1)
```

```python
for col in XT.columns:
    if XT[col].dtypes == 'object':
        encoder = LabelEncoder()
        XT[col] = encoder.fit_transform(XT[col])
```

```python
for col in XT.columns:
    scaler = StandardScaler()
    XT[col] = scaler.fit_transform(XT[col].values.reshape(-1, 1))
```

```python
dt = DecisionTreeClassifier(random_state=42)
```

```python
start_time = time.time()
dt.fit(X,y)
end_time = time.time()
training_time = end_time - start_time
print(f"Model training time: {training_time:.2f} seconds")
```

```python
yt = dft['income']
```

```python
dt_yprad = dt.predict(XT)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
dt_conf_mat = confusion_matrix(yt,dt_yprad)
print("confusion matrix for dt_model",'\n',dt_conf_mat)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
print('Decision Tree Classifier:')
print('Accuracy score:', round(accuracy_score(yt, dt_yprad) * 100, 2))
print('F1 score:', round(f1_score(yt, dt_yprad) * 100, 2))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
criterion=['gini', 'entropy']
max_depth = [int(x) for x in np.linspace(40, 150, num=15)]
```

```python
param_dist = {
    'criterion': criterion,
    'max_depth': max_depth}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_tuned = DecisionTreeClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf_tuned, param_grid=param_dist, cv=5, scoring='accuracy')
```

```python
start_time = time.time()
grid_search.fit(X, y)
end_time = time.time()
training_time = end_time - start_time
print(f"Model training time: {training_time:.2f} seconds")
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
grid_search_best = DecisionTreeClassifier(
    criterion='entropy', max_depth=40, random_state=42)
```

```python
Y_pred_rf_best = grid_search_best.predict(XT)
```

```python
print('Best Decision Tree Classifier:')
print('Accuracy score:', round(accuracy_score(yt, Y_pred_rf_best) * 100, 2))
print('F1 score:', round(f1_score(yt, Y_pred_rf_best) * 100, 2))
```
